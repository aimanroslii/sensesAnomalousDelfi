```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = pd.read_csv("x_train.csv")
x_valid = pd.read_csv("x_valid.csv")
x_test = pd.read_csv("x_test.csv")

y_train = pd.read_csv("y_train.csv")
y_valid = pd.read_csv("y_valid.csv")
y_test = pd.read_csv("y_test.csv")
```

```python
print(f"-X -Data Stats--")
print(f"Train Data      : {x_train.shape}")
print(f"Validation Data : {x_valid.shape}")
print(f"Test Data       : {x_test.shape}")
```

```python
print(f"-Y -Data Stats--")
print(f"Train Data      : {y_train.shape}")
print(f"Validation Data : {y_valid.shape}")
print(f"Test Data       : {y_test.shape}")
```

# Model Part

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn import datasets, metrics
```

```python
regressor = RandomForestRegressor(n_estimators = 100, random_state = 42)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_valid)
```

```python
y_pred = pd.DataFrame(y_pred, columns = ['PT','TT',"FT"])
```

```python
y_pred
```

```python
y_valid
```

```python
plt.figure(figsize = (20,10))
plt.scatter(y_valid, y_pred, color= 'red', label='Comparison of Prediction between Actual & Prediction Data')
plt.legend()
plt.grid()
plt.title('Random Forest Regression')
plt.xlabel('Prediction data')
plt.ylabel('Actual data')
plt.show()
```

# Accuracy Check

```python
metrics.r2_score(y_valid,y_pred)
```
